```python
x = torch.randn(10000, device="cuda" if torch.cuda.is_available() else "cpu")
```

```python
start_time = time.time()
_ = f(x)
eager_time = time.time() - start_time
print(f"Eager execution time: {eager_time} seconds")
```

```python
import torch._dynamo

# Allow torch.compile to fail and fall back to eager execution
torch._dynamo.config.suppress_errors = True

try:
    # Try to compile (Inductor backend by default)
    compiled_f = torch.compile(f)
    start_time = time.time()
    _ = compiled_f(x)
    graph_time = time.time() - start_time
    print(f"Graph execution time: {graph_time} seconds")

except Exception as e:
    # Absolute safety net: manual eager fallback
    print("Compilation failed, running eager instead")
    start_time = time.time()
    _ = f(x)
    graph_time = time.time() - start_time
    print(f"Eager execution time{graph_time} seconds")
```
